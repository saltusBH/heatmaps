```
import tensorflow as tf
import numpy as np
from tensorflow import keras
from tensorflow.contrib import lite

#モデル訓練
model = keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
model.compile(optimizer='sgd',loss='mean_squared_error')

xs = np.array([ -1.0,0.0,1.0,2.0,3.0,4.0],dtype=float)
ys = np.array([-3.0,-1.0,0.0,3.0,5.0,7.0],dtype=float)

model.fit(xs,ys,epochs=500)

#テスト結果表示
print(model.predict([10.0]))

#keras学習結果保存
keras_file = "linear.h5"
keras.models.save_model(model,keras_file)

#kerasファイルをtensorflow liteに変換
#converter = lite.TocoConverter.from_keras_model_file(keras_file)
converter = lite.TFLiteConverter.from_keras_model_file(keras_file)
tflite_model = converter.convert()
open("liner.tflite","wb").write(tflite_model)

#Googleドライブに出力
#upload_file_2 = drive.CreateFile()
#upload_file_2.SetContentFile("liner.tflite")
#upload_file_2.Upload()

from google.colab import files
files.download("linear.h5")
files.download("liner.tflite")
```
